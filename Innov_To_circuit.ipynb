```python
import csv
import ast
import pennylane as qml
```

```python
with open('FinalCircuit.csv') as innov: 
    rot = innov.readline()
    cnot = innov.readline()

x = rot.split('"')
del x[0]
rots = []
for i in range(len(x)):
    if (x[i] != ',' ):
        rots.append(x[i])

y = cnot.split('"')
del y[0]
cnots = []
for i in range(len(y)):
    if(y[i] != ','):
        cnots.append(y[i])

rot_dict = []
cnot_dict = []

for i in range(len(cnots)-1):
    temp1 = ast.literal_eval(cnots[i])
    cnot_dict.append(temp1)

for i in range(len(rots)-1):
    temp1 = ast.literal_eval(rots[i])
    rot_dict.append(temp1)

innov = cnot_dict + rot_dict

order = []

for i in range(len(innov)):
    if (innov[i]['layer'] == 0):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 1):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 2):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 3):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 4):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 5):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 6):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 7):
        order.append(innov[i])

for i in range(len(innov)):
    if (innov[i]['layer'] == 8):
        order.append(innov[i])

translate = []
for i in range(len(order)):
    if ('weights' in order[i]):
        temp = 'qml.Rot({}, {}, {}, wires={})'.format(order[i]['weights'][0], order[i]['weights'][1], order[i]['weights'][2], order[i]['wire'])
        translate.append(temp)
    elif ('to' in order[i]):
        translate.append('qml.CNOT(wires = [{fr}, {to}])'.format(fr = order[i]['wire'], to = order[i]['to']))
        #append CNOT to circuit
    
        
with open('circuit.txt', 'w') as f:
    for line in translate:
        f.write(line)
        f.write('\n')
```
